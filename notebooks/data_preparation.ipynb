```python
# Azure Demand Forecasting  Data Collection & Preparation



# Load datasets from data folder
raw = pd.read_csv("../data/raw_data.csv")
ext = pd.read_csv("../data/external_data.csv")

print("Raw Dataset:")
print(raw.head())

print("\nExternal Dataset:")
print(ext.head())

# Convert date columns
raw['date'] = pd.to_datetime(raw['date'])
ext['date'] = pd.to_datetime(ext['date'])

# Merge datasets
final = raw.merge(ext, on="date")

# Feature Engineering
final['month'] = final['date'].dt.month
final['year'] = final['date'].dt.year
final['quarter'] = final['date'].dt.quarter

print("\nMerged Dataset:")
print(final.head())

# Save processed dataset
final.to_csv("../data/processed_data.csv", index=False)

print("\n Processed dataset saved successfully!")
```
